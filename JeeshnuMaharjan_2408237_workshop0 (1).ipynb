# 4.1 Exercise on Functions (Workshop 0)
Task 1 — Unit Converter


```python
def convert_units():
    """
    Convert between different units of measurement.
    Supports length, weight and volume conversions.
    """

    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = int(input("Enter choice: "))
        value = float(input("Enter value to convert: "))

        if choice == 1:
            print("1. meters → feet")
            print("2. feet → meters")
            sub = int(input("Select option: "))

            if sub == 1:
                result = value * 3.28084
                print("Feet:", result)
            elif sub == 2:
                result = value / 3.28084
                print("Meters:", result)

        elif choice == 2:
            print("1. kg → lbs")
            print("2. lbs → kg")
            sub = int(input("Select option: "))

            if sub == 1:
                print("Pounds:", value * 2.20462)
            elif sub == 2:
                print("Kilograms:", value / 2.20462)

        elif choice == 3:
            print("1. liters → gallons")
            print("2. gallons → liters")
            sub = int(input("Select option: "))

            if sub == 1:
                print("Gallons:", value * 0.264172)
            elif sub == 2:
                print("Liters:", value / 0.264172)

        else:
            print("Invalid choice")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

convert_units()
```

Task 2 — Mathematical Operations on List

```python
def find_sum(numbers):
    return sum(numbers)

def find_average(numbers):
    return sum(numbers) / len(numbers)

def find_max(numbers):
    return max(numbers)

def find_min(numbers):
    return min(numbers)

try:
    nums = list(map(float, input("Enter numbers separated by space: ").split()))

    print("Choose operation:")
    print("1 Sum")
    print("2 Average")
    print("3 Maximum")
    print("4 Minimum")

    choice = int(input("Choice: "))

    if choice == 1:
        print("Sum:", find_sum(nums))
    elif choice == 2:
        print("Average:", find_average(nums))
    elif choice == 3:
        print("Max:", find_max(nums))
    elif choice == 4:
        print("Min:", find_min(nums))

except:
    print("Invalid input")
```

# Nested List Exercises
Sum of Nested List

```python
def sum_nested(lst):
    total = 0
    for i in lst:
        if isinstance(i, list):
            total += sum_nested(i)
        else:
            total += i
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

Remove Element

```python
def remove_element(lst, elem):
    result = []
    for i in lst:
        if isinstance(i, list):
            result.append(remove_element(i, elem))
        elif i != elem:
            result.append(i)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

Find Maximum

```python
def find_max(lst):
    maximum = float('-inf')

    for i in lst:
        if isinstance(i, list):
            maximum = max(maximum, find_max(i))
        else:
            maximum = max(maximum, i)

    return maximum

print(find_max([[1,2],[3,[4,5]],6]))
```

Count Occurrences

```python
def count_occurrences(lst, elem):
    count = 0

    for i in lst:
        if isinstance(i, list):
            count += count_occurrences(i, elem)
        elif i == elem:
            count += 1

    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

Flatten Nested List

```python
def deep_flatten(lst):
    result = []

    for i in lst:
        if isinstance(i, list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)

    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

Average of Nested List

```python
def average_nested(lst):

    flat = deep_flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
identity_check = np.round(np.dot(A, A_inv))
print("1. A × A⁻¹ = I (rounded):\n", identity_check)

AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("   BA:\n", BA)
print("   AB == BA?", np.array_equal(AB, BA))

AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)
print("\n3. (AB)^T:\n", AB_T)
print("   B^T × A^T:\n", BT_AT)
print("   Equal?", np.array_equal(AB_T, BT_AT))
```
